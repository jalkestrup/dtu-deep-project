## SEB Benchmark

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

with open(f"../seb_eval/seb_results_llama-8B-swe-simcse-dk.pkl", 'rb') as f:
    l2v_llama_swe_dk_results = pickle.load(f)

with open(f"../seb_eval/SEB_L2V_eval_LLM2Vec-Meta-Llama-3-8B-Instruct-mntp-unsup-simcse.pkl", 'rb') as f:
    l2v_llama_en_results = pickle.load(f)

task_names = [res.task_name for res in l2v_llama_swe_dk_results[0]]
task_names.append('Mean Score')
main_scores_llama_8B_swe_simcse_dk = [res.get_main_score() for res in l2v_llama_swe_dk_results[0]]
#Calculate and append the mean of the main scores for the swe simcse_dk model
mean_scores_llama_8B_swe_simcse_dk = np.mean(main_scores_llama_8B_swe_simcse_dk)
main_scores_llama_8B_swe_simcse_dk.append(mean_scores_llama_8B_swe_simcse_dk)

main_scores_llama_8B_en_simcse = [res.get_main_score() for res in l2v_llama_en_results[0]]
mean_scores_llama_8B_en_simcse = np.mean(main_scores_llama_8B_en_simcse)
main_scores_llama_8B_en_simcse.append(mean_scores_llama_8B_en_simcse)

fig, ax = plt.subplots(figsize=(10, 5))
bar_width = 0.35
index = np.arange(len(task_names))
ax.bar(index, main_scores_llama_8B_swe_simcse_dk, bar_width, label='Llama-8B-swe-simcse-ft-dk')
ax.bar(index + bar_width, main_scores_llama_8B_en_simcse, bar_width, label='Llama-8B-en-simcse')
ax.set_xticks(index + bar_width )
ax.set_xticklabels(task_names, rotation=90)
ax.set_ylabel('Primary Score')
ax.set_title('Scandinavian Embedding Benchmark, L2V Llama EN vs. L2V Llama-swe DK finetuned')
ax.legend()
plt.show()
```
