```python
import seb

all_tasks = seb.get_all_tasks()
#Create a map of name and task type by irterating over all tasks
task_type_map = {}
for task in all_tasks:
    task_type_map[task.name] = task.task_type

task_type_map
```

```python
category_count = {
    "Classification": 7,
    "BitextMining": 1,
    "Retrieval": 3
}

category_scores = {}
for label, filepath in files_and_labels.items():
    with open(filepath, 'rb') as f:
        results = pickle.load(f)
    # Create the outer dictionary with the label as the key
    category_scores[label] = {}
    for res in results[0][:-1]:
        task_type = task_type_map[res.task_name]
        if task_type not in category_scores[label]:
            category_scores[label][task_type] = []
        category_scores[label][task_type].append(res.get_main_score())  

    #Calculate the mean score for each category and overwrite the list of scores with the mean score
    for task_type, scores in category_scores[label].items():
        mean_score = np.mean(scores)
        category_scores[label][task_type] = mean_score
        
category_scores
```
